### Description of the Game:
Cluedo, also known as Clue in North America, is a classic murder mystery board game where players assume the roles of characters trying to solve a murder. The game is set in a mansion, and players move around to different rooms, gather clues, and make deductions to identify the murderer, the weapon used, and the room where the crime occurred. ### Clue rules
* Objective: be first player to figure out who committed the crime, with what happen, and in what location
* Randomly select a combination of player, weapon, and room cards and place in envelop
* Deal out remaining cards to all players
* Remainder of cards are placed face up next to the board
* Roll dice, and move character that many places othogonally on the board (no diagnoal moves)
* If land of red question mark, you may do what it says if you want to look
* Players may not share places in the corridor
* If you enter a room you must make a suggestion
* i.e., "green with the lamp in the bedroom"
* The player to the left (or right?) secretely shows you one card that matches the description
    * If no cards match, the next player secretly shows you, etc
* You can make only 1 accusation per game, where you go to the center and check if match
    * if correct, you win
    * if not, you lose, but still participate in the remainder of the game

## Rules of Clue

1. Game Setup
* Characters: 6 suspects.
* Weapons: 6 weapons.
* Rooms: 9 rooms.
* Murder Envelope: One random suspect, one random weapon, and one random room are secretly chosen — the solution.
* Distribution: The remaining cards are shuffled and dealt to all players.
* Player Start: Each player picks a suspect and starts near the board’s edge or center depending on your rule variation.
* All characters and weapons start in the center (even unused ones).
* Select 1 random suspect, weapon, and room into the secret envelope.
* Shuffle and deal the remaining cards evenly to all players.
* Uneven extras go face up beside the board.

Players take private notes based on what they’re dealt and shown.
3. Turns and Movement
* Each turn consists of:
* Rolling a die (or choosing a move if not rolling).
* Moving through the hallways to enter rooms (1 step per square).
* Moving into a room only via doorways.
* If a player enters a room, they may immediately make a suggestion.
* Roll dice and move orthogonally (not diagonally).
* Optional: draw a bonus card when landing on a red “?” tile.
* If in a room, may:
* Stay
* Use a secret passage
* Or walk out through any door

You must stop upon entering a room.

4. Suggestions (In a Room)
* Player suggests:
* “I think it was Professor Plum, in the Library, with the Candlestick.”
* The suggested suspect and weapon are moved into that room.
* In clockwise order, other players must reveal one matching card if they have any.
* Only the suggester sees the revealed card.

5. Accusations (Anywhere)
* Once per game, a player can make an accusation (same format).
    * If correct: they win.
    * If wrong: they’re out — but continue revealing cards for suggestions.


6. Card Dealing & Deduction
* Cards are split among players.
* Players track suggestions, responses, and revealed cards to deduce the hidden solution.

```python
class Room:
    def __init__(self, name, board_layout):
        self.name = name
        self.room_designation = name.lower().replace(" ", "_")
        self.cells = self.get_cells(board_layout)
        self.entrances = self.get_entrance(board_layout)

    def get_cells(self, board_layout):
        """Return list of (row, col) tuples where this room exists on the board."""
        coords = []
        for r in range(board_layout.shape[0]):
            for c in range(board_layout.shape[1]):
                val = board_layout.iat[r, c]
                if isinstance(val, str) and val == self.room_designation:
                    coords.append((r, c))
        return coords

    def get_entrance(self, board_layout):
        """Return list of (row, col) for room entrances (e.g., 'study_e')."""
        coords = []
        entrance_label = self.room_designation + "_e"
        for r in range(board_layout.shape[0]):
            for c in range(board_layout.shape[1]):
                val = board_layout.iat[r, c]
                if isinstance(val, str) and val == entrance_label:
                    coords.append((r, c))
        return coords

    def __repr__(self):
        return f"{self.name} (cells: {len(self.cells)}, entrances: {self.entrances})"
```
